```python
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
dataset = dataset.astype(int)
dataset
```

```python
dataset = dataset.drop("User ID",axis=1)
dataset
```

```python
independent = dataset[['Age',	'EstimatedSalary', 'Gender_Male']]
independent
```

```python
dependent = dataset[["Purchased"]]
dependent
```

```python
from mlstats import Classification as classifier

acc_table, max_acc, min_acc, best_param = classifier.knn(independent, dependent)
acc_table
print(best_param)
```

```python
acc_table, max_acc, min_acc, best_param = classifier.decision_tree(independent, dependent)
acc_table
print(best_param)
```

```python
acc_table, max_acc, min_acc, best_param = classifier.random_forest(independent, dependent)
acc_table
print(best_param)
```

```python
classif_report = classifier.fit(independent, dependent)
classif_report
```

```python
classif_report = classifier.fit_save(independent, dependent)
```
